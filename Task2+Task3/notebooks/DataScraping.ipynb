```python
import yfinance as yf
import pandas as pd

def fetch_history(ticker, period="1year", interval="1d"):
    t = yf.Ticker(ticker)
    hist = t.history(period=period, interval=interval, actions = False)
    return hist

# Quick run
hist = fetch_history("AAPL", period = "1y", interval = "1d")
hist.head()
```
